```python
for link in links:
    print('===============================')
    print(link.get('href'))
    print('================================')
    
print('Total number of links is: ',len(links))
```

```python
movies = []
movies_title = []
movies_links = []
```

```python
td_data = soup.find_all('a')
for item in td_data:
    movies.append(item.text)
```

```python
for item in td_data:
    movies_title.append(item.get('title'))
```

```python
for item in td_data:
    movies_links.append(item.get('href'))
```
